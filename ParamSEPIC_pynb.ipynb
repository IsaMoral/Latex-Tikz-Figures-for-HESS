```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros del convertidor SEPIC
Vout = 12  # Voltaje de entrada (V)
fsw = 60e3  # Frecuencia de conmutación (Hz)
R = 10  # Resistencia de carga (Ω)

# Rango de voltajes de salida
Vin_range = np.arange(0, 50, 0.2)

# Inicializar listas para almacenar los valores de ciclo de trabajo
duty_cycle = []

# Calcular el ciclo de trabajo para cada voltaje de salida
for Vin in Vin_range:
    # Ecuación para el ciclo de trabajo en modo de conducción continua
    D = (Vout / Vin) / (1 + (Vout / Vin))
    duty_cycle.append(D)

# Crear el gráfico
plt.figure(figsize=(10, 6))
plt.plot(Vin_range, duty_cycle)
plt.title('Ciclo de trabajo vs. Voltaje de salida (Convertidor SEPIC)')
plt.xlabel('Voltaje de Entrada (V)')
plt.ylabel('Ciclo de trabajo (D)')
plt.grid(True)
plt.show()
```
